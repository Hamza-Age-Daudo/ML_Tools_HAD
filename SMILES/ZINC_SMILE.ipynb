```python
import pandas as pd
import requests
import time
from google.colab import files

# Função para obter SMILES a partir de um ID ZINC
def get_smiles(zinc_id):
    url = f'http://zinc15.docking.org/substances/{zinc_id}.json'
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Levanta um erro para status de resposta 4xx/5xx
        data = response.json()
        if 'smiles' in data:
            return data['smiles']
        else:
            return 'SMILES não encontrado'
    except requests.exceptions.RequestException as req_err:
        return f'Error: {req_err}'

# Passo 1: Importar o arquivo CSV
uploaded = files.upload()

# Ler o arquivo CSV que foi carregado
for filename in uploaded.keys():
    print(f'Arquivo carregado: {filename}')
    df = pd.read_csv(filename)
    break  # Supondo que você tenha apenas um arquivo CSV carregado

# Exibir as primeiras linhas do DataFrame para verificar
print(df.head())

# Passo 2: Obter SMILES e CID para cada ID ZINC
zinc_ids = df['Name'].tolist()  # Substitua 'ZINC_ID' pelo nome correto da coluna

# Dicionário para armazenar os resultados
results_dict = {'Name': [], 'SMILES': []}

for zinc_id in zinc_ids:
    smiles = get_smiles(zinc_id)

    results_dict['Name'].append(zinc_id)
    results_dict['SMILES'].append(smiles)

# Criar um DataFrame com os resultados
results_df = pd.DataFrame(results_dict)

# Salvar os resultados em um arquivo CSV
results_file = 'zinc_pubchem_results.csv'
results_df.to_csv(results_file, index=False)

# Oferecer o arquivo para download
files.download(results_file)
```
